```python
from sklearn.linear_model import Ridge

rr = Ridge(alpha=.1)
```

```python
predictors = data.columns[~data.columns.isin(["target", "name", "station"])]
```

```python
def backtest(weather, model, predictors, start=3050, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(data, rr, predictors)
```

```python
predictions
```

```python
#accuracy metrics
from sklearn.metrics import mean_absolute_error
mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
#other accuracy mertric less is better
predictions["diff"].mean()
```

```python
#percentage difference
def pct_diff(old, new):
    return (new - old) / old
#finding rolling avg of a certain period
#rolling(horizon) for finding mean of rows
def compute_rolling(data, horizon, col):
    label = f"rolling_{horizon}_{col}"
    data[label] = data[col].rolling(horizon).mean()
    data[f"{label}_pct"] = pct_diff(data[label], data[col])
    return data

rolling_horizons = [3, 14]  #it takes from previous 14 days now
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        data = compute_rolling(data, horizon, col)
```

```python
data = data.iloc[14:,:]
data
```

```python
data=data.fillna(0)
```

```python
#taking month of every year as every year the season would be same
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    data[f"month_avg_{col}"] = data[col].groupby(data.index.month, group_keys=False).apply(expand_mean)
    data[f"day_avg_{col}"] = data[col].groupby(data.index.day_of_year, group_keys=False).apply(expand_mean)
```
